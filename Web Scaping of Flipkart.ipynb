```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
import re
import requests
```

```python
phn_name = []
phn_price = []
phn_stars = []
phn_image = []
phn_other_details = []

page_num = int(input("Enter number of pages: "))
for i in range(1,page_num+1):
    url = "https://www.flipkart.com/search?q=mi+mobiles&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_3_na_na_na&as-pos=1&as-type=RECENT&suggestionId=mi+mobiles%7CMobiles&requestId=a5cdf79c-1f21-4132-9311-ac3d439f9315&as-searchtext=mi+&page=2"+str(i)
    req = requests.get(url)
    content = bs(req.content,'html.parser')
    names = content.find_all("div", {"class": "_4rR01T"})
    prices = content.find_all("div",{"class": "_30jeq3 _1_WHN1"})
    stars =  content.find_all("div",{"class": "_3LWZlK"})
    images = content.find_all("div",{"class": "CXW8mj"})
    other_details = content.find_all("li",{"class": "rgWa7D"})
    
    for name in names:
        phn_name.append(name.get_text())
    
    for price in prices:
        phn_price.append(price.get_text())

    for star in stars:
        phn_stars.append(star.get_text())

    for image in images:
        phn_image.append(image.img['src'])
        
    for other_detail in other_details:
        phn_other_details.append(other_detail.get_text())
    
```

```python
ram = []
display = []
camera = []
battery = []
warrenty = []
j=1
k=1
t=1
for i in phn_other_details:
    print(t,end=" ")
    print(i)
    if(j==1):
        ram.append(i)
        j= j+1
        continue
    if(j==2):
        display.append(i)
        j = j+1
        continue
    if(j==3):
        camera.append(i)
        j = j+1
        continue
    if(j==4):
        battery.append(i)
        j= j+1
        continue
    if "Brand" in i:
        k=2
    elif "Warranty" in i:
        k=2
    elif "warranty" in i:
        k = 2
    elif "Year" in i:
        k = 2
    elif "year" in i:
        k=2
    elif "months" in i:
        k=2
    elif "Months" in i:
        k=2
    elif "YEAR" in i:
        k=2
    
    if(j==5 and k==2):
        warrenty.append(i)
        t=t+1
        print(t,end=" ")
        print(i)
        j=1
        k=1
```

```python
data = {'Phone_name': phn_name,'Phone_price': phn_price,'Phone_ram':ram,'Phone_display':display,'Phone_camera':camera,"Phone_battery":battery,"Phone_Warrenty":warrenty}
```

```python
df = pd.DataFrame(data)
df = df.reset_index(drop = True)
print(df)
```

```python
df.to_csv("Records.csv",index = False)
```
